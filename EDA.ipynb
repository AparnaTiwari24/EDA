```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\Aparna Tiwari\Downloads\train_and_test2.csv")
```

```python
df.head()
```

```python
df = df.rename(columns ={'2urvived':'Survived'})
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
print(df[['Sex', 'Pclass', 'Embarked', 'Survived']].value_counts())
```

```python
df['Age'].hist(bins=30, edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
df['Fare'].hist(bins=30, edgecolor='black')
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.countplot(x='Sex',data=df)
plt.title("Gender Distribution")
plt.show()
```

```python
sns.countplot(x='Pclass', data=df)
plt.title("Passenger Class Distribution")
plt.show()
```

```python
sns.countplot(x='Survived',data =df)
plt.title("Survived Count")
plt.show()
```

```python
sns.boxplot(x='Survived',y='Age', data =df)
plt.title("Age vs Survival")
plt.show()
```

```python
sns.boxplot(x='Pclass',y='Fare',data=df)
plt.title("Fare by Passenger Class")
plt.show()
```

```python
sns.pairplot(df[['Age','Fare','Pclass','Survived']].dropna(),hue='Survived')
plt.show()
```

```python
corr = df[['Age','Fare','Pclass','Survived']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
print("Observations:")
print("- Most passengers were in 3rd class.")
print("- Majority did not survive (~74%).")
print("- Females (encoded as 1) had a higher survival rate.")
print("- Higher fare and younger age seem positively related to survival.")
```
